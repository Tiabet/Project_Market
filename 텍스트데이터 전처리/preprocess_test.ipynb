```python
import csv
import pandas as pd
import unicodedata

tsv_file = '/content/drive/MyDrive/강이지_옷/강아지옷.tsv'

data = []

with open(tsv_file, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    for row in reader:
        data.append(row)

df = pd.DataFrame(data[1:], columns=data[0])

df = df[:len(df)//100]

df
```

```python
#전각문자 제거

import unicodedata

reviews = df['review']

for review in df['review']:
  review = unicodedata.normalize('NFKC', review)

reviews
```

```python
# 띄어쓰기, 맞춤법 정정

from hanspell import spell_checker
import re

spell_checked = []

for review in reviews:
  spelled_sent = spell_checker.check(review)
  text = spelled_sent.checked
  spell_checked.append(text)
```

```python
df_preprocessed = pd.DataFrame({'original': reviews, 'spell_checked': spell_checked})
```

```python
# 성능에 일부 문제가 있는 부분이 있어서
# 띄어쓰기, 맞춤법 재정정

spell_checked_again = []

for review in df_preprocessed['spell_checked']:
  spelled_sent = spell_checker.check(review)
  text = spelled_sent.checked
  spell_checked_again.append(text)

df_preprocessed['spell_checked'] = spell_checked_again
```

```python
# 정규식 사용
# 1. 초성 제거
# 2. 특수문자 제거 (괄호는 제거하지 않음)
# 3. :) :( 이모티콘 제거 후 : 도 제거
# 4. kg, ml 등 영어 단위 제거
# 5. 두 칸 이상의 여백, 탭 등 남아 있을 수도 있는 불필요한 띄어쓰기 제거

re_checked = []

for text in df_preprocessed['spell_checked']:
  only_BMP_pattern = re.compile("["
        u"\U00010000-\U0010FFFF"  #BMP characters 이외
                           "]+", flags=re.UNICODE)
  text = only_BMP_pattern.sub(r'', text)
  text = re.sub(r'[ㄱ-ㅎㅏ-ㅣ0-9]+', '', text)
  text = re.sub('ᄒ+', '', text)
  text = re.sub('[ෆ⃛❤❤❤♥♡】૮₍˶•⑅₎ა]', '', text)
  text = re.sub('[-=+,#/\?^.@*\";※~ㆍ!』‘|\<\>\[\]\_`\'…》\”\“\’·]', ' ', text)
  text = re.sub(r':[)D]|:[(]','', text)
  text = re.sub(r':','', text)
  text = re.sub(r'[a-zA-Z]{1,2}', '', text)
  text = re.sub(r'\s{2,}|\t', ' ', text)
  re_checked.append(text)
```

```python
df_preprocessed['re+spell_checked'] = re_checked
df_preprocessed
```
