```python
import pandas as pd
from collections import Counter
from nltk.util import ngrams
from konlpy.tag import Okt
import csv
```

```python
tsv_file = '테스트용_preprocessed.tsv'

data = []

with open(tsv_file, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    for row in reader:
        data.append(row)

df = pd.DataFrame(data[1:], columns=data[0])
df
```

```python
stopwords_txt = 'stopwords.txt'

with open(stopwords_txt, 'r', encoding='utf-8') as file:
    stopwords= [line.strip() for line in file.readlines()]
custom_stop_words = ["강아지","구매","감사","생각","우리","아기","주문","너무","잘","거","같아요","더"]

stopwords.extend(custom_stop_words)
```

```python
combined_text = ' '.join(df['review'])

words = combined_text.split()

filtered_words = [word for word in words if word not in stopwords]

n=1

n_grams = list(ngrams(filtered_words, n))
ngram_counts = Counter(n_grams)

print("Most common", n, "grams from the combined text:")
for ngram, count in ngram_counts.most_common():
    if count > len(df['review'])//20:
        print(' '.join(ngram), "-", count)
```
