```python
import csv
import pandas as pd
import unicodedata

# Replace 'data.tsv' with the path to your TSV file
tsv_file = '/content/drive/MyDrive/강이지_옷/강아지옷.tsv'

# Initialize an empty list to store the data from the TSV file
data = []

# Read the TSV file and store its content in the 'data' list
with open(tsv_file, 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter='\t')
    for row in reader:
        data.append(row)

df = pd.DataFrame(data[1:], columns=data[0])

df = df[:len(df)//100]

df
```

```python
#전각문자 제거

import unicodedata

reviews = df['review']

for review in df['review']:
  review = unicodedata.normalize('NFKC', review)

reviews
```

```python
# 띄어쓰기, 맞춤법 정정

from hanspell import spell_checker
import re

spell_checked = []

for review in reviews:
  spelled_sent = spell_checker.check(review)
  hanspell_sent = spelled_sent.checked
  spell_checked.append(hanspell_sent)
```

```python
df_preprocessed = pd.DataFrame({'original': reviews, 'spell_checked': spell_checked})
```

```python
# 성능에 일부 문제가 있는 부분이 있어서
# 띄어쓰기, 맞춤법 재정정

spell_checked_again = []

for review in df_preprocessed['spell_checked']:
  spelled_sent = spell_checker.check(review)
  hanspell_sent = spelled_sent.checked
  spell_checked_again.append(hanspell_sent)

df_preprocessed['spell_checked'] = spell_checked_again
```

```python
# 정규식 사용
# 1. 초성 제거
# 2. 특수문자 제거
# 3. kg, ml 등 영어 단위 제거
# 4. 두 칸 이상의 여백, 탭 등 남아 있을 수도 있는 불필요한 띄어쓰기 제거

re_checked = []

for hanspell_sent in df_preprocessed['spell_checked']:
  #hangul = re.compile('[^ 가-힣 a-zA-Z()]')
  #hanspell_sent = hangul.sub('', hanspell_sent)
  hanspell_sent = re.sub(r'[ㄱ-ㅎㅏ-ㅣ0-9]+', '', hanspell_sent)
  hanspell_sent = re.sub('[-=+,#/\?:^.@*\"※~ㆍ!』‘|\[\]`\'…》\”\“\’·]', ' ', hanspell_sent)
  hanspell_sent = re.sub(r'[a-zA-Z]{1,2}', '', hanspell_sent)
  hanspell_sent = re.sub(r'\s{2,}|\t', ' ', hanspell_sent)
  re_checked.append(hanspell_sent)
```

```python
df_preprocessed['re+spell_checked'] = re_checked
df_preprocessed
```
